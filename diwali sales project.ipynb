# ***EDA***


```python
df.columns
```

```python
cntplt=sns.countplot(x='Gender',data=df)
for bars in cntplt.containers:
    cntplt.bar_label(bars)
    plt.title("count graph")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# to get the amount of male and female
g_Amount= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
g_Amount
```

```python

sns.barplot(x='Gender',y='Amount',data=g_Amount)
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Bar graph")
plt.show()
```

**most of the buyer are female and their purchasing worth is also high**

```python
gend_Grph= sns.countplot(x='Age Group',data=df,hue="Gender")
for bars in gend_Grph.containers:
    gend_Grph.bar_label(bars)
plt.show()
```

##state 

```python
df.columns
```

```python
t_ord= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x="State",y="Orders",data=t_ord)

plt.show()
```

```python
t_wohrt= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

t_wohrt


```

```python
plt.figure(figsize=(20,5))

sns.barplot(x="State",y="Amount",data=t_wohrt)
plt.show()
```

```python
df.columns
```

```python
m_dta=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in m_dta.containers:
    m_dta.bar_label(bars)
plt.show()
```

```python
m_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=m_status,hue="Gender")
plt.title('Marital_status VS Amount')
plt.show()
```

from above most of the buyer are non married women and their purchasing power is also high.
